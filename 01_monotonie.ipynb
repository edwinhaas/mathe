```python
import sympy as sy
from sympy.plotting import plot
```

```python
x,y = sy.symbols("x, y")
```

Funktion $f(x) = 4x + x^2$

```python
f = 4*x+x*x
```

Ableitung $f'(x) = \frac{df(x)}{dx} $

```python
f1 = sy.diff(f,x)
f1
```

Zeichnen wir beide Funktionen, $f(x)$ und darunter $f'(x)$

```python
p1 = plot(f, show=False)
p2 = plot(f1, show=False)
p1.show()
p2.show()
```

Monotonie dannst Du nun aus der zweiten Abbildung ablesen. Die zweite Abiildung $f'(x)$ 
zeigt die Steigung der Funktion $f(x)$ an. Wenn $f'(x) > 0 $ ist, bedeutet dass das die Funktion $f(x)$ immer weiter ansteigt. Nimmst Du einen Punkt $x_1 \in X$ und einen zweiten Punkt $x_2 \in X$ mit $x_1 < x_2$ dann muss $f(x_1) < f(x_2) $ 

Eine Besonderheit stellt der Wert $x=2$ dar, denn an diesem Punkt hat die Funktion $f(x)$ keine Steigung, sie verläuft waagerecht. 

Links von diesem Punkt verhält sich alles entgegengesetzt. Die Funktion $f(x)$ ist monoton fallend, wenn man sich von weit links an den Wert $x=2$ herantastet, dann gilt 
$f(x_1) > f(x_2) $ für $x_1 < x_2$ 

Insgesamt kannst Du folgendes aussagen:
* $f(x)$ ist nicht monoton, denn es findet einen wechsel im Vorzeichen der Funktion * $f'(x)$ statt, und zwar an der Stelle $x=2$  
* $f(x)$ ist im Abschnitt $x > 2$ monoton steigend und
* $f(x)$ ist im Abschnitt $x < 2$ monoton fallend
